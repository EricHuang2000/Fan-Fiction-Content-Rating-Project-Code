```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("C:/Users/erich/Desktop/DS_project/data/comp.xlsx")

algorithms = data['algri'].unique()
metrics = data['Metric'].unique()

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for i, metric in enumerate(metrics):
    ax = axes[i]
    for alg in algorithms:
        subset = data[(data['Metric'] == metric) & (data['algri'] == alg)]
        ax.plot(subset['number'], subset['value'], label=alg, marker='o')
        
    ax.set_title(metric)
    ax.set_xlabel('Number of Sentences')
    ax.set_ylabel('Value')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    "Harry felt he is flying when he saw the magic shown to him",
    "James feels danger is coming but it when he is flying",
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print("Original Text:", corpus)
print("After Tokenization:",vectorizer.get_feature_names())
print("Vecotrs:",X.toarray())
```
